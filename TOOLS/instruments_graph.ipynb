```python
from anytree import Node, RenderTree
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Axel\Documents\PYTHON\MSP_V3\FragHub\datas\GRAPH_CONSTRUCT_SOLUTIONS.xlsx", sheet_name="AB SCIEX")
df = df.astype(str)
df_dict = df.to_dict(orient='records')
```

```python
trees = {}

for row in df_dict:
    if row['MARQUES'] not in trees:
        trees[row['MARQUES']] = Node(row['MARQUES'])
    node = trees[row['MARQUES']]
    for col in ['MODELS', 'SPECTRUM_TYPE', 'INSTRUMENT_TYPE', 'IONISATION', 'RESOLUTION', 'SOLUTION']:
        try:
            node = [child for child in node.children if child.name == row[col]][0]
        except IndexError:
            node = Node(row[col], parent=node)
```

```python
for tree in trees.values():
    for pre, fill, node in RenderTree(tree):
        print("%s%s" % (pre, node.name))
```
